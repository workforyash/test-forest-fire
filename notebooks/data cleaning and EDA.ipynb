## Performing EDA on the cleaned data

```python
df_copy= df.copy()
```

```python
df_copy.head()
```

```python
#droping day, month and year as they are not the factors that affect the temperature
df_copy = df_copy.drop(labels=['day', 'month', 'year'], axis=1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
#converting the classes column into numerical with 0 and 1 being not fire and fire
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
#ploting density function for all features
df_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# get the percentage of classes in the data
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
#pie chart for categorical attribute
classLabels=["Fire", "Not Fire"]
plt.figure(figsize=(12,12))
plt.pie(percentage, labels=classLabels, autopct='%1.1f%%' )
plt.show()
```

```python
#correlation heatmap
plt.figure(figsize=(20, 10))
sns.heatmap(df_copy.corr(), annot=True)
```
